```python
#Beginning: Download latest data from https://github.com/CSSEGISandData/COVID-19/tree/master/csse_covid_19_data/csse_covid_19_time_series 

import pandas as pd 

confirmed_raw = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv", index_col = 'Country/Region')
confirmed = pd.DataFrame(confirmed_raw.groupby('Country/Region').sum())
confirmed = confirmed.sort_values(by='3/15/20')

deaths_raw = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv", index_col = 'Country/Region')
deaths = pd.DataFrame(deaths_raw.groupby('Country/Region').sum())

recovered_raw = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv", index_col = 'Country/Region')
recovered = pd.DataFrame(recovered_raw.groupby('Country/Region').sum())
```

```python
selection1 = confirmed.loc[['Germany','France','Spain','Austria','Switzerland','UK','Poland','Denmark','US','Pakistan','Egypt','Korea, South']]
cf1 = selection1.transpose().tail(15)
#cf1
#cf11 = cf1.sort_values(by='3/15/20')
 
```

```python
cf1.plot.line(title='Confirmed1')
```

```python
selection2 = confirmed.loc[['Germany','France','UK','Spain','Italy','China','Iran','Korea, South','Japan','Taiwan*','India','Pakistan']]
cf2 = selection2.transpose().tail(30)

cf2.plot.line(title='Confirmed2')
```

```python
selection3 = deaths.loc[['Germany','Italy','Spain','Iran','China','Japan','Taiwan','India','Pakistan','Korea, South']]
ds1 = selection3.transpose().tail(30)

ds1.plot.line(title='Deaths')
```
