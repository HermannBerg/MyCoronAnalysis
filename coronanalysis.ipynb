```python
#Beginning: Download latest data from https://github.com/CSSEGISandData/COVID-19/tree/master/csse_covid_19_data/csse_covid_19_time_series 


import pandas as pd 

confirmed_raw = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv", index_col = 'Country/Region')
confirmed = pd.DataFrame(confirmed_raw.groupby('Country/Region').sum())
confirmed = confirmed.sort_values(by='3/15/20')

deaths_raw = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv", index_col = 'Country/Region')
deaths = pd.DataFrame(deaths_raw.groupby('Country/Region').sum())

#recovered_raw = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv", index_col = 'Country/Region')
#recovered = pd.DataFrame(recovered_raw.groupby('Country/Region').sum())

confirmed.to_csv('confirmed.csv')
deaths.to_csv('deaths.csv')
#recovered.to_csv('recovered.csv')
```
