```python
#Beginning: Download latest data from https://github.com/CSSEGISandData/COVID-19/tree/master/csse_covid_19_data/csse_covid_19_time_series

import pandas as pd 

confirmed = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv", index_col = 'Country/Region')
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv", index_col = 'Country/Region')
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv", index_col = 'Country/Region')
```

```python
covid_frame = pd.DataFrame(confirmed)

Selection = covid_frame.loc[['Germany','Italy','Iran','Japan','Taiwan','India']]

SelectionT = Selection.transpose()
SelectionT.plot.line(title='Confirmed')

```

```python
Selection2 = covid_frame.loc[['Germany','France','Spain','Austria','Switzerland','UK','Poland','Denmark']]

SelectionT2 = Selection2.transpose()
SelectionT2.plot.line(title='Confirmed')
```

```python
covid_frame_deaths = pd.DataFrame(deaths)

SelectionDeaths = covid_frame_deaths.loc[['Germany','Italy','Iran','Japan','Taiwan','India']]

SelectionDeathsT = SelectionDeaths.transpose()
SelectionDeathsT.plot.line(title='Deaths')
```

```python
covid_frame_deaths = pd.DataFrame(deaths)

SelectionDeaths = covid_frame_deaths.loc[['Germany','France','Spain','Austria','Switzerland','UK','Poland','Denmark']]

SelectionDeathsT = SelectionDeaths.transpose()
SelectionDeathsT.plot.line(title='Deaths')
```

```python
covid_frame_recovered = pd.DataFrame(recovered)

SelectionRecovered = covid_frame_recovered.loc[['Germany','France','Spain','Austria','Switzerland','UK','Poland','Denmark']]

SelectionRecoveredT = SelectionRecovered.transpose()
SelectionRecoveredT.plot.line(title='Recovered')
```
